***Preprocessing the coin***





```python
import pandas as pd


coins = ['BTC', 'ETH', 'DOGE', 'AAVE', 'PURR/USDC', 'kPEPE', '@107', 'BRETT', 'MOODENG']

majors = {'BTC','ETH','SOL','BNB','XRP','ADA','DOGE','LTC','DOT','MATIC','TRX','ATOM'}
meme = {'SHIB','PEPE','DOGE','WIF','BONK','FLOKI','BRETT','POPCAT','MEME','MOODENG','BOME','FARTCOIN'}
stable_pairs = {'USDT','USDC','DAI','USTC','PAXG'}
# altcoins = everything else (but valid tickers)
# internal = tokens starting with "@"
# wrapped = tokens starting with "k"

def categorize_coin(coin):
    if coin in majors:
        return "Major"
    elif coin in meme:
        return "Meme"
    elif any(stable in coin for stable in stable_pairs) or "/" in coin:
        return "Stable/Pair"
    elif coin.startswith("@"):
        return "Internal/Unknown"
    elif coin.startswith("k"):
        return "Wrapped/Prefixed"
    else:
        return "Altcoin/Other"

historical_data["Coin_Category"] = historical_data["Coin"].apply(categorize_coin)
historical_data.head()
```

# ***correlations between sentiment indicators and trading outcomes (PnL, win rate, risk, profit factor, Sharpe ratio, maximum drawdown, trading frequency, ROI, etc.)***

```python
sentiment_mapping = {
    "Extreme Fear": -2,
    "Fear": -1,
    "Neutral": 0,
    "Greed": 1,
    "Extreme Greed": 2
}

all_merged_data['sentiment_score'] = all_merged_data['classification'].map(sentiment_mapping)
```

```python
columns_of_interest = ['sentiment_score', 'value', 'Net PnL','Size USD' ,'Size Tokens','Execution Price','Start Position']
corr_df = all_merged_data[columns_of_interest].corr()

sns.heatmap(corr_df, annot=True, cmap='crest')
plt.show()

```

***findings :***



*   Sentiment and value move together (very strong correlation).
*   Better sentiment/value → slightly better Net PnL.
*   Trade size (USD) and execution price are connected.(Trade Size (USD)       =Execution PricexSize Tokens)
*   Size Tokens and Start Position are weakly related.



Weighted Win Rate = Winning Trade Size (Tokens) / Total Trade Size (Tokens)

Win Rate = Number of Winning Trades / Total Number of Trades

ROI = (Net Profit (PnL) ÷ Investment Amount) x 100


Max Drawdown=max(Peak Value−Current Value)


```python

summary_df = all_merged_data.groupby('classification').agg(

    win_rate=('Net PnL', lambda x: (x > 0).sum() / len(x)),


    win_rate_weighted=('Size Tokens', lambda x: x[all_merged_data.loc[x.index, 'Net PnL'] > 0].sum() / x.sum()),


    avg_net_pnl=('Net PnL', 'mean'),


    total_net_pnl=('Net PnL', 'sum'),

    trade_count=('Net PnL', 'count'),


    avg_roi=('Net PnL', lambda x: (x / all_merged_data.loc[x.index, 'Size USD']).mean() * 100)
).reset_index()

summary_df = summary_df.sort_values(by='win_rate_weighted', ascending=False)

summary_df
```

***Take Away:***




*   Greed → frequent and large profitable trades → safest/best sentiment to trade.
*   Extreme Greed → fewer winning trades, but high ROI per winning trade.
*   Fear / Extreme Fear / Neutral → low or negative profitability → high-risk sentiment.



```python

metrics_df = all_merged_data.groupby('classification').agg(
    closed_pnl_sum=('Closed PnL', 'sum'),
    closed_pnl_avg=('Closed PnL', 'mean'),
    gross_profit=('Net PnL', lambda x: x[x > 0].sum()),
    gross_loss=('Net PnL', lambda x: -x[x < 0].sum()),
    roi=('Size USD', lambda x: (all_merged_data.loc[x.index, 'Net PnL'].sum() / x.sum()) if x.sum() != 0 else np.nan)
).reset_index()


metrics_df['profit_factor'] = metrics_df['gross_profit'] / metrics_df['gross_loss']


metrics_df['sharpe_ratio'] = all_merged_data.groupby('classification')['Net PnL'].transform(
    lambda x: x.mean() / (x.std() if x.std() != 0 else 1)
).groupby(all_merged_data['classification']).first().values


def max_drawdown(series):
    cumulative = series.cumsum()
    peak = cumulative.cummax()
    drawdown = peak - cumulative
    return drawdown.max()

max_dd = all_merged_data.groupby('classification')['Net PnL'].apply(max_drawdown)
metrics_df = metrics_df.merge(max_dd.rename('max_drawdown'), on='classification')


metrics_df = metrics_df.sort_values(by='roi', ascending=False)

metrics_df
```

***findings :***



*   Extreme Greed → small number of trades but extremely profitable per trade, very low drawdown.

*   Greed → many profitable trades but higher drawdown.

*   Fear / Extreme Fear / Neutral → mostly losing trades, low or negative ROI, moderate risk.



# ***Generate ranked lists of traders based on performance under specific sentiment classes or trend types***

```python


trader_sentiment_df = all_merged_data.groupby(['classification', 'Account']).agg(
    total_trades=('Trade ID', 'count'),
    net_pnl_sum=('Net PnL', 'sum'),
    net_pnl_avg=('Net PnL', 'mean'),
    win_rate=('Net PnL', lambda x: (x > 0).sum() / len(x)),
    win_rate_weighted=('Size Tokens', lambda x: x[all_merged_data.loc[x.index, 'Net PnL'] > 0].sum() / x.sum()
                       if x.sum() != 0 else np.nan),
    gross_profit=('Net PnL', lambda x: x[x > 0].sum()),
    gross_loss=('Net PnL', lambda x: -x[x < 0].sum()),
).reset_index()


trader_sentiment_df['profit_factor'] = trader_sentiment_df['gross_profit'] / trader_sentiment_df['gross_loss'].replace(0, np.nan)

roi_df = all_merged_data.groupby(['classification', 'Account'])[['Net PnL', 'Size USD']].apply(
    lambda x: x['Net PnL'].sum() / x['Size USD'].sum() if x['Size USD'].sum() != 0 else np.nan
).reset_index(name='roi')



trader_sentiment_df = trader_sentiment_df.merge(roi_df, on=['classification','Account'])


trader_sentiment_df['rank_by_pnl'] = trader_sentiment_df.groupby('classification')['net_pnl_sum'] \
                                                        .rank(method='dense', ascending=False)

trader_sentiment_df['rank_by_roi'] = trader_sentiment_df.groupby('classification')['roi'] \
                                                       .rank(method='dense', ascending=False)


trader_sentiment_df = trader_sentiment_df.sort_values(['classification','rank_by_pnl'])

trader_sentiment_df.head()

```
